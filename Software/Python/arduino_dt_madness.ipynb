```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
stream = open('stream3.txt', 'r')

times = []
for line in stream.readlines():
    if line.startswith('T'):
        _, t = line.split(':')
        times.append(int(t))
```

```python
times = np.array(times)
dts = times[1:] - times[:-1]
```

```python
plt.figure()

plt.semilogx(times[:-1], dts)
```
